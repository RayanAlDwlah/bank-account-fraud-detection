# **Encoding**

# **Feature Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train = pd.DataFrame(X_train, columns=cols)
X_test = pd.DataFrame(X_test, columns=cols)
```

# **Feature Selection**

```python
from sklearn.ensemble import ExtraTreesClassifier


model = ExtraTreesClassifier()
model.fit(X_train, y_train)

feat_importances = pd.Series(model.feature_importances_, index=X_train.columns)

plt.figure(figsize=(12, 15)) 
feat_importances.nlargest(48).sort_values().plot(kind='barh', color='skyblue')
plt.title('Top 48 Features by Importance (ExtraTreesClassifier)')
plt.xlabel('Feature Importance')
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, chi2

selector = SelectKBest(score_func=chi2, k='all')
selector.fit(X_train, y_train)

scores = selector.scores_
feature_scores = pd.Series(scores, index=X_train.columns)
top_features = feature_scores.nlargest(42)  

plt.figure(figsize=(10,8))
top_features.sort_values().plot(kind='barh')
plt.title('Top Features by Chi2 Score')
plt.xlabel('Chi2 Score')
plt.ylabel('Features')
plt.grid(True)
plt.show()
```

```python
k = 29  

selector = SelectKBest(score_func=chi2, k=k)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

selected_features = X_train.columns[selector.get_support()]
print("Selected features:\n", selected_features.tolist())

X_train_selected = pd.DataFrame(X_train_selected, columns=selected_features)
X_test_selected = pd.DataFrame(X_test_selected, columns=selected_features)

print("X_train_selected shape:", X_train_selected.shape)
print("X_test_selected shape:", X_test_selected.shape)
```

```python
X_train_selected.head()
```

# **Resampling**

```python
from imblearn.combine import SMOTEENN
from collections import Counter

smote_enn = SMOTEENN(random_state=42)

X_resampled, y_resampled = smote_enn.fit_resample(X_train_selected, y_train)

print("Original class distribution:", Counter(y_train))
print("Resampled class distribution:", Counter(y_resampled))

print("X_resampled shape:", X_resampled.shape)
print("y_resampled shape:", y_resampled.shape)
```

 # **Models and Hyperparameters**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

models = {
    'Logistic Regression': LogisticRegression(max_iter=500, class_weight='balanced'),
    'Random Forest': RandomForestClassifier(class_weight='balanced'),
    'AdaBoost': AdaBoostClassifier(),
    'Bagging': BaggingClassifier(
    estimator=DecisionTreeClassifier(class_weight='balanced'),
    n_jobs=-1),  
    'Gradient Boosting': GradientBoostingClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

params = {
    'Logistic Regression': {'C': [0.001,0.00001,0.01, 0.1, 1, 10, 100, 1000]},
    'Random Forest': {'n_estimators': [100], 'max_depth': [7], 'min_samples_split': [2]},
    'AdaBoost': {'n_estimators': [ 200], 'learning_rate': [ 1]},

    'Bagging': {'n_estimators': [ 100]},
    'Gradient Boosting': {'n_estimators': [50, 100, 200], 'learning_rate': [0.1, 0.01], 'max_depth': [3, 5, 7]},
    'XGBoost': {'n_estimators': [100, 200, 300], 'learning_rate': [0.01, 0.1], 'max_depth': [3, 6, 9]}
}

best_models = {}

for model_name, model in models.items():
    print(f"Processing Model: {model_name}...")
    grid_search = GridSearchCV(estimator=model,
                            param_grid=params.get(model_name, {}),
                            cv=cv,
                            scoring='accuracy',
                            n_jobs=-1,
                            verbose=1)
    grid_search.fit(X_resampled, y_resampled)
    best_models[model_name] = {
        'Best Estimator': grid_search.best_estimator_,
        'Best Parameters': grid_search.best_params_,
        'Best Score': grid_search.best_score_
    }
    print(f"{model_name}:\n Best Parameters: {grid_search.best_params_}\n Best Score: {grid_search.best_score_}\n")
```

# **Training Best Models & Evaluating Accuracy**

```python
from sklearn.metrics import accuracy_score

for name, model_dict in best_models.items():
    model = model_dict['Best Estimator']  

    model.fit(X_resampled, y_resampled)
    
    train_pred = model.predict(X_resampled)
    test_pred = model.predict(X_test_selected)
    
    train_acc = accuracy_score(y_resampled, train_pred)
    test_acc = accuracy_score(y_test, test_pred)
    
    print(f"🔹 {name}:\n- Train Accuracy: {train_acc:.4f}\n- Test Accuracy: {test_acc:.4f}\n{'-'*50}")
```

```python
from sklearn.metrics import (
    confusion_matrix, classification_report,
    roc_auc_score, roc_curve, accuracy_score
)

def evaluate_model(name, y_true, y_pred, y_proba):
    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    roc_auc = roc_auc_score(y_true, y_proba)
    plt.figure(figsize=(7, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {name}')
    plt.legend(loc='lower right')
    plt.show()

    # Classification Report
    print(f"\n📊 Classification Report for {name}:")
    print(classification_report(y_true, y_pred))
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print('-' * 60)
```

```python
test_predictions = {}

for name, model_dict in best_models.items():
    model = model_dict['Best Estimator']
    y_pred = model.predict(X_test_selected)
    y_proba = model.predict_proba(X_test_selected)[:, 1]
    test_predictions[name] = (y_pred, y_proba)
    evaluate_model(name, y_test, y_pred, y_proba)
```

# **Voting Classifier**

```python
from sklearn.ensemble import VotingClassifier

voting_clf = VotingClassifier(
    estimators=[
        ('xgb', best_models['XGBoost']['Best Estimator']),
        ('gb', best_models['Gradient Boosting']['Best Estimator']),
        ('lr', best_models['Logistic Regression']['Best Estimator'])
    ],
    voting='soft'
)

voting_clf.fit(X_resampled, y_resampled)

y_pred_voting = voting_clf.predict(X_test_selected)
y_proba_voting = voting_clf.predict_proba(X_test_selected)[:, 1]

evaluate_model("Voting Classifier", y_test, y_pred_voting, y_proba_voting)
```

# **»Stacking Classifier**

```python
from sklearn.ensemble import StackingClassifier

stacking_clf = StackingClassifier(
    estimators=[
        ('xgb', best_models['XGBoost']['Best Estimator']),
        ('gb', best_models['Gradient Boosting']['Best Estimator']),
        ('ada', best_models['AdaBoost']['Best Estimator']),
    ],
    final_estimator=LogisticRegression()
)
stacking_clf.fit(X_resampled, y_resampled)

y_pred_stack = stacking_clf.predict(X_test_selected)
y_proba_stack = stacking_clf.predict_proba(X_test_selected)[:, 1]

evaluate_model("Stacking Classifier", y_test, y_pred_stack, y_proba_stack)
```

# **Custom Threshold Evaluation (Recall @ 5% FPR)**

```python
from sklearn.metrics import roc_curve, classification_report, confusion_matrix, accuracy_score, roc_auc_score

def evaluate_custom_threshold_model(name, y_true, y_proba, threshold):
    y_pred = (y_proba >= threshold).astype(int)

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name} (Threshold = {threshold:.4f})')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    fpr, tpr, _ = roc_curve(y_true, y_proba)
    roc_auc = roc_auc_score(y_true, y_proba)
    plt.figure(figsize=(7, 5))
    plt.plot(fpr, tpr, color='orange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.title(f'ROC Curve - {name}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.show()

    print(f"\n📊 Classification Report for {name}:")
    print(classification_report(y_true, y_pred))
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print('-'*60)
```

# **Apply Threshold @ 5% FPR**

```python
print("🔎 Evaluation using Recall @ 5% FPR")
print("="*60)

for name, (y_pred, y_proba) in test_predictions.items():
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    target_fpr = 0.05
    optimal_idx = np.argmin(np.abs(fpr - target_fpr))
    optimal_threshold = thresholds[optimal_idx]

    evaluate_custom_threshold_model(f"{name} (Threshold @ 5% FPR)", y_test, y_proba, optimal_threshold)
```
